```python
# paso 2)se carga la data
dataset = pd.read_csv('https://raw.githubusercontent.com/mk-gurucharan/Regression/master/SampleData.csv')
X = dataset.iloc[:, 0].values
y = dataset.iloc[:, 1].values
y = np.array(y).reshape(-1,1)
dataset.head(5)
```

```python
#Paso 3) se escala la data para que trabaje mejor el modelo



from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()
sc_y = StandardScaler()
X = sc_X.fit_transform(X.reshape(-1,1))
y = sc_y.fit_transform(y.reshape(-1,1))
y()
```

```python
#Paso 4) test set & training set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

```python
#paso 5 entrenando SVR en el training set

from sklearn.svm import SVR
regressor = SVR(kernel = 'rbf')
regressor.fit(X_train.reshape(-1,1), y_train.reshape(-1,1))
```

```python
#paso 6 predicción del test set
y_pred = regressor.predict(X_test)
y_pred = np.array(y_pred).reshape(-1,1)
y_pred = sc_y.inverse_transform(y_pred)
y_pred
```

```python
from sklearn.metrics import accuracy_score #para ver la precisión del modelo
print(accuracy_score(y_test,y_pred))
```
